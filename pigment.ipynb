# Fitting (Not Log10)

```python
def fit_and_plot_2(M, xdata, ydata, x_range_plot, sigma, linewidth=1, color=-1, linestyle='--'):
    fitObj = CurveFit_2(xdata, ydata, sigma)
    fitParams, fitCovariance = fitObj.fit(M)
    chi2 = fitObj.calcChi2()  
    
    print M, '(fitting) :'
    print '    Params:           ', fitParams
    print '    Params Covariance:', fitCovariance
    print '    Chi2             :', chi2
    
    yfit = fitObj.fitfunction(x_range_plot, fitParams)
    if color!=-1:
        plt.plot(x_range_plot, yfit, linestyle=linestyle, label='fit %d' % M, linewidth=linewidth, color=color)
    else:
        plt.plot(x_range_plot, yfit, linestyle=linestyle, label='fit %d' % M, linewidth=linewidth)
#     if color==-1:
#         plt.plot(x_range_plot, yfit, ls='--', label='fit %d' % M, linewidth=linewidth, color=color)
#     if linewidth==-1:
#         plt.plot(x_range_plot, yfit, ls='--', label='fit %d' % M, linewidth=linewidth, color=color)
#     if linewidth==-1:
#         plt.plot(x_range_plot, yfit, ls='--', label='fit %d' % M, linewidth=linewidth, color=color)
```
